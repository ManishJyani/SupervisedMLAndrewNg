In this lab, you will contrast regression and classification.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([0., 1, 2, 3, 4, 5]) #single variable input
y_train = np.array([0,  0, 0, 1, 1, 1])
X_train2 = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]]) # 2 variable input
y_train2 = np.array([0, 0, 0, 1, 1, 1])
```

```python
pos=y_train==1
neg= y_train==0

pos2= y_train2==1
neg2= y_train2==0
fig, ax = plt.subplots(1,2,figsize=(10,3))
ax[0].scatter(x_train[pos],y_train[pos],c='b',marker='o',label='y=1')
ax[0].scatter(x_train[neg],y_train[neg],c='r',marker='x',label='y=0')
ax[0].set_title("One Varialble")
ax[0].set_xlabel('X')
ax[0].set_ylabel("Y")
ax[0].legend()
ax[1].scatter(X_train2[pos2][:,0],X_train2[pos2][:,1],c='b',marker='o',label='y=1')
ax[1].scatter(X_train2[neg2][:,0],X_train2[neg2][:,1],c='r',marker='x',label='y=0')
ax[1].set_title("Two Varialble")
ax[1].set_xlabel("$X_1$")
ax[1].set_ylabel("$X_2$")
plt.tight_layout()
ax[1].legend()

# fig.show()
```
