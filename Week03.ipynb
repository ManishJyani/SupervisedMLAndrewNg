# C1_W3_Lab04_LogisticLoss_Soln
In this ungraded lab, you will:
- explore the reason the squared error loss is not appropriate for logistic regression
- explore the logistic loss function

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x_train = np.array([0., 1, 2, 3, 4, 5],dtype=np.longdouble)
y_train = np.array([0,  0, 0, 1, 1, 1],dtype=np.longdouble)
pos= y_train==1
neg= y_train==0

plt.scatter(x_train[pos],y_train[pos],c='r',marker='x',label='malignant')
plt.scatter(x_train[neg],y_train[neg],c='b',marker='o',label='benign')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
```

```python
## let's try to plot the squared error function (which was cost function in regression problem)
```

Now, let's get a surface plot of the cost using a *squared error cost*:
  $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2 $$ 
 
where 
  $$f_{w,b}(x^{(i)}) = sigmoid(wx^{(i)} + b )$$

```python
# if you plot cost function against weights and bias , then it willl be wiggly and so many local minima will be there that you can not find optimum weights and bias and this
# method will fail to serve the purpose , so this cost function won't work , we need something convax type 
```

Logistic Regression uses a loss function more suited to the task of categorization where the target is 0 or 1 rather than any number. 

>**Definition Note:**   In this course, these definitions are used:  
**Loss** is a measure of the difference of a single example to its target value while the  
**Cost** is a measure of the losses over the training set


This is defined: 
* $loss(f_{\mathbf{w},b}(\mathbf{x}^{(i)}), y^{(i)})$ is the cost for a single data point, which is:

\begin{equation}
  loss(f_{\mathbf{w},b}(\mathbf{x}^{(i)}), y^{(i)}) = \begin{cases}
    - \log\left(f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right) & \text{if $y^{(i)}=1$}\\
    - \log \left( 1 - f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right) & \text{if $y^{(i)}=0$}
  \end{cases}
\end{equation}


*  $f_{\mathbf{w},b}(\mathbf{x}^{(i)})$ is the model's prediction, while $y^{(i)}$ is the target value.

*  $f_{\mathbf{w},b}(\mathbf{x}^{(i)}) = g(\mathbf{w} \cdot\mathbf{x}^{(i)}+b)$ where function $g$ is the sigmoid function.
* Notational convention: `log` means the natural logarithm. 

The defining feature of this loss function is the fact that it uses two separate curves. One for the case when the target is zero or ($y=0$) and another for when the target is one ($y=1$). Combined, these curves provide the behavior useful for a loss function, namely, being zero when the prediction matches the target and rapidly increasing in value as the prediction differs from the target. # C1_W3_Lab05_Cost_Function_Soln
## Goals
In this lab, we will:
- examine the implementation and utilize the cost function for logistic regression.

```python
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])  #(m,n)
y_train = np.array([0, 0, 0, 1, 1, 1])                                           #(m,)
pos= y_train==1
neg= y_train==0
```

```python
fig,ax= plt.subplots(1,1,figsize=(4,4))
ax.scatter(X_train[:,0][pos],X_train[:,1][pos],c='r',marker='x',label='maligant')
ax.scatter(X_train[:,0][neg],X_train[:,1][neg],c='b',marker='o',label='benign')
plt.xlabel("$X_1$"),
ax.set_ylabel("$X_2$")
plt.legend()
```

```python
# compute the logistic loss function
def sigmoid(x,w,b):
    z= np.dot(x,w)+b
    return 1/(1+math.exp(-z))  # this sigmoid is referred as f(w,b)
```

```python
# cost funtion= -y*log(f(w,b))-(1-y)*log(1-f(w,b)) ,this is good way to combine the cost function for both y=0 and y=1
def compute_cost_logistic(X,y,w,b):
    m,n= X.shape
    loss=0
    for i in range(m):
        f= sigmoid(X[i],w,b)
     
        loss= loss + ( -y[i]*np.log(f)-(1-y[i])*np.log(1-f) )
    


    cost = loss/m
    return cost

```

```python
w_tmp = np.array([1,1])
b_tmp = -3
print(compute_cost_logistic(X_train, y_train, w_tmp, b_tmp))
```

```python
# now just taking a random set and plotting decesion boundary
import matplotlib.pyplot as plt

# Choose values between 0 and 6
x0 = np.arange(0,6)

# Plot the two decision boundaries
x1 = 3 - x0
x1_other = 4 - x0

fig,ax = plt.subplots(1, 1, figsize=(4,4))
# Plot the decision boundary
ax.plot(x0,x1, c='b', label="$b$=-3")
ax.plot(x0,x1_other, c='k', label="$b$=-4")
ax.axis([0, 4, 0, 4])

# Plot the original data
pos= y_train==1
neg=y_train==0
ax.scatter(X_train[:,0][pos],X_train[:,1][pos],c='r',marker='x',label='maligant')
ax.scatter(X_train[:,0][neg],X_train[:,1][neg],c='b',marker='o',label='benign')
ax.axis([0, 4, 0, 4])
ax.set_ylabel('$x_1$', fontsize=12)
ax.set_xlabel('$x_0$', fontsize=12)
plt.legend(loc="upper right")
plt.title("Decision Boundary")
plt.show()
```

```python
# so b= -3 is giving better result ,we can caculate cost function for it
w_array1 = np.array([1,1])
b_1 = -3
w_array2 = np.array([1,1])
b_2 = -4

print("Cost for b = -3 : ", compute_cost_logistic(X_train, y_train, w_array1, b_1))
print("Cost for b = -4 : ", compute_cost_logistic(X_train, y_train, w_array2, b_2))
```
